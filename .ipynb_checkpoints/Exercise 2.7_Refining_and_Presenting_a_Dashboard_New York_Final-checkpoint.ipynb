# Exercise 2.7_Refining and Presenting a Dashboard_New York_Latest

```python
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from datetime import datetime as dt
import folium
```

```python
df = pd.read_csv("New York_Weather_trip_data_Final.csv")
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df['started_at'] = pd.to_datetime(df['started_at'])
df['ended_at'] = pd.to_datetime(df['ended_at'])
```

```python
## Groupby

df['value'] = 1 
df_groupby_bar = df.groupby('start_station_name', as_index=False).agg({'value': 'sum'})
top20 = df_groupby_bar.nlargest(20, 'value')
```

```python
## Groupby

df['value'] = 1 
df_groupby_bar1 = df.groupby('end_station_name', as_index=False).agg({'value': 'sum'})
top20 = df_groupby_bar1.nlargest(20, 'value')
```

```python
df_groupby_bar = df_groupby_bar.rename(columns={
    "start_station_name": "station",
})

df_groupby_bar1 = df_groupby_bar1.rename(columns={
    "end_station_name": "station",
})
```

```python
df_groupby_bar["station"] = df_groupby_bar["station"].str.strip().str.lower()
df_groupby_bar1["station"] = df_groupby_bar1["station"].str.strip().str.lower()
```

```python
top_20_latest = (
    pd.concat([df_groupby_bar, df_groupby_bar1])
    .groupby("station", as_index=False)["value"]
    .sum()
    .sort_values("value", ascending=False)
    .head(20)
)
top_20_latest['usage_type'] = 'High Demand'
top_20_latest['recommendation'] = 'Increase bike/dock capacity, frequent rebalancing, prioritize maintenance'
top_20_latest.to_csv("New York_Weather_trip_data_Updated_top20.csv")
```

```python
# Plot the bar chart for the top 20 popular stations
fig = go.Figure(
    go.Bar(
        x=top_20_latest['value'],          # numeric values on x-axis
        y=top_20_latest['station'],        # categories on y-axis
        orientation='h',                   # makes it horizontal
        marker=dict(
            color=top_20_latest['value'],
            colorscale='Blues'
        )
    )
)

fig.update_layout(
    title=dict(
        text="Top 20 Most Popular Bike Stations in New York",
        x=0.5,
        xanchor='center'
    ),
    width=1200,
    height=800,
    xaxis_title="Sum of Trips",
    yaxis_title="Station Names",
    yaxis=dict(autorange="reversed"),
    margin=dict(l=250)   # important for long station names
)

fig.show()
```

```python
least_20_latest = (
    pd.concat([df_groupby_bar, df_groupby_bar1])
    .groupby("station", as_index=False)["value"]
    .sum()
    .sort_values("value", ascending=False)
    .tail(20)
)
least_20_latest['usage_type'] = 'Low Demand'
least_20_latest['recommendation'] = 'Consider reducing bikes/docks, promotional incentives, evaluate location'
least_20_latest.to_csv("New York_Weather_trip_data_Updated_least20.csv")
```

```python
# Plot the bar chart for least 20 popular stations in New York

fig = go.Figure(
    go.Bar(
        x=least_20_latest['value'],          # numeric values on x-axis
        y=least_20_latest['station'],        # categories on y-axis
        orientation='h',                   # makes it horizontal
        marker=dict(
            color=least_20_latest['value'],
            colorscale='Blues'
        )
    )
)

fig.update_layout(
    title=dict(
        text="Top 20 Least Popular Bike Stations in New York",
        x=0.5,
        xanchor='center'
    ),
    width=1200,
    height=800,
    xaxis_title="Sum of Trips",
    yaxis_title="Station Names",
    yaxis=dict(autorange="reversed"),
    margin=dict(l=250)   # important for long station names
)

fig.show()
```

### Line Chart for Average Temperature Vs Daily Ride count

```python
# Groupby the count of the trips 

df_group = pd.DataFrame(df.groupby(['date','season'])['ride_id'].count()).reset_index()
```

```python
df_group.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)
```

```python
# Group the records by date and its average temperature

df_group1 = pd.DataFrame(df.groupby(['date','season'])['avgTemp'].mean()).reset_index()
```

```python
# Merge the dataframes with the daily ride count and average temperature
df_merged = df_group.merge(
    df_group1[['date', 'avgTemp']],
    on='date',
    how='left'
)
```

```python
df_merged
```

```python
df_sorted = df_merged.sort_values('date')
season_transitions = df_sorted[df_sorted['season'] != df_sorted['season'].shift(1)]
transition_dates = season_transitions[['date', 'season']]
```

```python
df_sorted
```

```python
df_sorted.to_csv("New York_Weather_trip_data_Updated_AvgTemp.csv")
```

```python
# Plot the line chart with Average Temperature Vs Daily Ride count

fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Scatter(
        x=df_sorted['date'],
        y=df_sorted['bike_rides_daily'],
        name='Daily bike rides',
        line=dict(color='steelblue')
    ),
    secondary_y=False
)

fig.add_trace(
    go.Scatter(
        x=df_sorted['date'],
        y=df_sorted['avgTemp'],
        name='Daily temperature',
        line=dict(color='firebrick')
    ),
    secondary_y=True
)

# Title
fig.update_layout(
    title='Daily Rides and Temperature Over Time'
)

fig.update_xaxes(title_text='Date')

fig.update_yaxes(
    title_text='Number of Bike Rides',
    secondary_y=False
)

fig.update_yaxes(
    title_text='Average Temperature (C)',
    secondary_y=True
)

for _, row in transition_dates.iterrows():
    fig.add_shape(
        type="line",
        x0=row['date'],
        x1=row['date'],
        y0=0,
        y1=1,
        xref='x',
        yref='paper',  # 0 to 1 of y-axis (spans entire height)
        line=dict(color='gray', width=1, dash='dash')
    )
    
    # optional annotation
    fig.add_annotation(
        x=row['date'],
        y=1.02,  # slightly above the plot
        xref='x',
        yref='paper',
        text=row['season'].capitalize(),
        showarrow=False,
        align='center'
    )
fig.show()
```

### Line Chart with Membership Type Vs Daily ride count

```python
# Extract 'hour' value from the started_at column
df['ride_hour'] = df['started_at'].dt.hour
```

```python
# Peak hours by member type
member_hourly = (
    df.groupby(['member_casual', 'ride_hour'])
      .size()
      .reset_index(name='total_rides')
)
member_hourly.to_csv("New York_Weather_trip_data_membership_type_ride_count.csv")
```

```python
member_hourly.shape
```

```python
# Create an empty figure
fig = go.Figure()

# Add one line per membership type
for member_type in member_hourly['member_casual'].unique():
    subset = member_hourly[member_hourly['member_casual'] == member_type]
    fig.add_trace(go.Scatter(
        x=subset['ride_hour'],
        y=subset['total_rides'],
        mode='lines+markers',  # lines with markers
        name=member_type
    ))

# Update layout
fig.update_layout(
    title='Peak Riding Hours by Membership Type',
    xaxis_title='Hour of Day (023)',
    yaxis_title='Number of Rides',
    xaxis=dict(tickmode='linear', tick0=0, dtick=1),  # ticks 023
    template='plotly_white'
)

# Show interactive figure in scripts or Jupyter
fig.show()
```

```python
# Peak hour vs Day of the week
df['day_of_week'] = df['started_at'].dt.day_name()
```

```python
# Calculate the Trip duration in minutes using the started_at and ended_at time.
df['trip_duration_min'] = (df['ended_at'] - df['started_at']).dt.total_seconds()/60
```

```python
df = df[df['trip_duration_min'] > 0]
df = df[df['trip_duration_min'] <= df['trip_duration_min'].quantile(0.99)]
```

```python
import pandas as pd
import numpy as np

def calculate_box_stats(df):
    """
    Calculate Q1, median, Q3, lowerfence, upperfence
    per member_casual group for boxplot plotting.
    """
    
    summary_list = []
    
    # Group by user type
    grouped = df.groupby('member_casual')['trip_duration_min']
    
    for user_type, data in grouped:
        data = data.dropna()  # remove missing values
        
        q1 = np.percentile(data, 25)
        median = np.percentile(data, 50)
        q3 = np.percentile(data, 75)
        
        iqr = q3 - q1
        
        # Tukey fences
        lowerfence = max(data.min(), q1 - 1.5 * iqr)
        upperfence = min(data.max(), q3 + 1.5 * iqr)
        
        summary_list.append({
            "user_type": user_type,
            "q1": q1,
            "median": median,
            "q3": q3,
            "lowerfence": lowerfence,
            "upperfence": upperfence
        })
    
    summary_df = pd.DataFrame(summary_list)
    
    return summary_df
```

```python
summary_df = calculate_box_stats(df)

print(summary_df)
```

```python
summary_df.to_csv("NewYork_Weather_trip_data_boxplot_stats.csv", index=False)
```

```python
# Separate stats
member_stats = summary_df[summary_df['user_type'] == 'member'].iloc[0]
casual_stats = summary_df[summary_df['user_type'] == 'casual'].iloc[0]

# Determine common y-axis range
y_min = min(member_stats["lowerfence"], casual_stats["lowerfence"])
y_max = max(member_stats["upperfence"], casual_stats["upperfence"])

# Create subplots: 1 row, 2 columns
fig = make_subplots(rows=1, cols=2, subplot_titles=("Member", "Casual"))

# Member box
fig.add_trace(go.Box(
    q1=[member_stats["q1"]],
    median=[member_stats["median"]],
    q3=[member_stats["q3"]],
    lowerfence=[member_stats["lowerfence"]],
    upperfence=[member_stats["upperfence"]],
    name="Member",
    marker_color="royalblue",
    boxmean=False,  # hide mean
    showlegend=False
), row=1, col=1)

# Add median annotation for Member
fig.add_annotation(
    x=0,  # first subplot
    y=member_stats["median"],
    text=f"Median: {member_stats['median']:.1f}",
    showarrow=True,
    arrowhead=2,
    row=1,
    col=1
)

# Casual box
fig.add_trace(go.Box(
    q1=[casual_stats["q1"]],
    median=[casual_stats["median"]],
    q3=[casual_stats["q3"]],
    lowerfence=[casual_stats["lowerfence"]],
    upperfence=[casual_stats["upperfence"]],
    name="Casual",
    marker_color="red",
    boxmean=False,
    showlegend=False
), row=1, col=2)

# Add median annotation for Casual
fig.add_annotation(
    x=1,  # second subplot
    y=casual_stats["median"],
    text=f"Median: {casual_stats['median']:.1f}",
    showarrow=True,
    arrowhead=2,
    row=1,
    col=2
)

# Update layout
fig.update_layout(
    title_text="Trip Duration (Minutes) by User Type",
    template="plotly_white",
    width=1000,
    height=500
)

# Add axis labels and enforce same y-axis
fig.update_xaxes(title_text="User Type", row=1, col=1)
fig.update_xaxes(title_text="User Type", row=1, col=2)
fig.update_yaxes(title_text="Trip Duration (Minutes)", row=1, col=1, range=[y_min, y_max])
fig.update_yaxes(title_text="Trip Duration (Minutes)", row=1, col=2, range=[y_min, y_max])

# Show figure
fig.show()
```

### Plot the bar chart for rides per day and hour of the day

```python
rides_per_hour_df = df.groupby('ride_hour', as_index=False).agg(total_rides=('ride_id', 'count'))
                                                                    
rides_per_day_df = df.groupby('day_of_week', as_index=False).agg(total_rides=('ride_id', 'count'))
```

```python
rides_per_hour_df.to_csv("NewYork_Weather_trip_data_rides_per_hour.csv", index=False)
rides_per_day_df.to_csv("NewYork_Weather_trip_data_rides_per_day.csv", index=False)
```

```python

fig = make_subplots(rows=1, cols=2,
                    subplot_titles=("Ride Frequency by Hour",
                                    "Ride Frequency by Day"))

# --- Rides by Hour ---
fig.add_trace(
    go.Bar(
        x=rides_per_hour_df['ride_hour'],
        y=rides_per_hour_df['total_rides'],
        name="Hour"
    ),
    row=1, col=1
)

# --- Rides by Day ---
fig.add_trace(
    go.Bar(
        x=rides_per_day_df['day_of_week'],
        y=rides_per_day_df['total_rides'],
        name="Day"
    ),
    row=1, col=2
)

fig.update_layout(
    height=500,
    width=1000,
    showlegend=False,
    template="plotly_white",
    title_text="Ride Analysis Dashboard"
)
# --- X-axis labels ---
fig.update_xaxes(title_text="Hour of Day", row=1, col=1)
fig.update_xaxes(title_text="Day of Week", row=1, col=2)

# --- Y-axis labels ---
fig.update_yaxes(title_text="Total Rides", row=1, col=1)
fig.update_yaxes(title_text="Total Rides", row=1, col=2)


fig.show()

```

```python
# Bar chart to plot the ride type usage by membership type
counts = (
    df.groupby(['member_casual', 'rideable_type'])
      .size()
      .unstack()
)
```

```python
counts
```

```python
import plotly.express as px

# Reset index so 'member_casual' becomes a column
df_plotly = counts.reset_index()

# Convert wide to long format for Plotly
df_long = df_plotly.melt(
    id_vars='member_casual',
    var_name='rideable_type',
    value_name='number_of_rides'
)

# Define custom colors for each rideable type
rideable_colors = {
    'classic_bike': 'royalblue',
    'electric_bike': 'orange',
    'docked_bike': 'green'
    # Add more rideable types if needed
}

# Create Plotly bar chart with custom colors
fig = px.bar(
    df_long,
    x='member_casual',
    y='number_of_rides',
    color='rideable_type',
    barmode='group',  # side-by-side bars
    labels={
        'member_casual': 'Membership Type',
        'number_of_rides': 'Number of Rides',
        'rideable_type': 'Ride Type'
    },
    title='Ride Type Usage by Membership Type',
    color_discrete_map=rideable_colors  # Apply custom colors
)

fig.show()
```
